```python
# Añado una columna que contenga un único string con toda la info de localización:

df2['Address'] = df2['Location'].astype(str) + ', ' + \
                df2['Area'] + ', ' + \
                df2['Country']

display(df2.Address[0])
display(df2.Address[1])
```

```python
# Defino una función que me devuelva una localización menos específica,
# eliminando la primera parte del string antes de la coma:

def cut(x):
    x = x.split(', ', maxsplit=1)[1:]
    return x[0]
```

```python

print(df2.Address[0])
print(df2.Address[1])# es una string

prueba = df2.Address[0]
prueba1 = df2.Address[1]# la acorto

print('\n')
print(cut(prueba))
print(cut(prueba1))# me devuelve otra string
```

```python
# Quiero crear una función que dé una string de Address al geocoder:

from time import sleep #Gracias Fran!!
import geopandas as gpd
import geopy
from geopy.geocoders import Nominatim

# Defino la función aplicable a cada string en 'address':
# Gracias Clara!!

def coord(x):
    sleep(1)
    geolocator = Nominatim(user_agent = "myGeocoder")
    location = geolocator.geocode(x)
    if location == None:
        plan_b = cut(x)
        location = geolocator.geocode(plan_b)
        return(location.latitude, location.longitude)
    else:
        return(location.latitude, location.longitude)
    
```

```python
df2['Point'] = df2.Address.apply(coord)

display(df2)
```

```python
# Convierto 'point' en coordenadas de 'Latitud' y 'Longitud'
df2[['Latitud', 'Longitud']] = pd.DataFrame(df2['Point'].tolist(), index=df2.index)
df2
```
